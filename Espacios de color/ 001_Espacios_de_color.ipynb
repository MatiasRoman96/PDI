Ejercicio 1: Espacios de Color

Usando una imagen a color de su elección:

a) Cargar la imagen con OpenCV y mostrar los canales BGR por separado.
b) Identificar cuál de los tres canales tiene mayor información basándose en los valores promedio.
c) Convertir la imagen de BGR a RGB y explicar por qué los colores se ven diferentes antes y después de la conversión.

```python
#instalar y cargar librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Cargar la imagen desde la ruta
img = cv2.imread('/content/palta.jpg')

# Verificar si se cargó correctamente
if img is None:
    print("La imagen no se pudo cargar. Verificá la ruta.")
else:
    # Mostrar imagen tal cual en BGR (esto se verá con colores alterados)
    plt.imshow(img)
    plt.title('Imagen en BGR (vista incorrecta con Matplotlib)')
    plt.axis('off')
    plt.show()

    # Separar canales BGR
    b, g, r = cv2.split(img)

    # Mostrar canal azul
    plt.imshow(b, cmap='gray')
    plt.title('Canal Azul')
    plt.axis('off')
    plt.show()

    # Mostrar canal verde
    plt.imshow(g, cmap='gray')
    plt.title('Canal Verde')
    plt.axis('off')
    plt.show()

    # Mostrar canal rojo
    plt.imshow(r, cmap='gray')
    plt.title('Canal Rojo')
    plt.axis('off')
    plt.show()
```

```python
# Calcular promedios de intensidad
avg_b = np.mean(b)
avg_g = np.mean(g)
avg_r = np.mean(r)

print(f"Promedio Canal Azul: {avg_b:.2f}")
print(f"Promedio Canal Verde: {avg_g:.2f}")
print(f"Promedio Canal Rojo: {avg_r:.2f}")

# Determinar cuál tiene más información
canales = {'Azul': avg_b, 'Verde': avg_g, 'Rojo': avg_r}
canal_mayor_info = max(canales, key=canales.get)
print(f"\n➡️ El canal con más información (mayor valor promedio) es: {canal_mayor_info}")
print("Valores cercanos a 0 indican ausencia del color y cercanos a 255 indican máxima intensidad")
```

```python
# Convertir imagen de BGR a RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Mostrar imagen en RGB
plt.imshow(img_rgb)
plt.title('Imagen en RGB')
plt.axis('off')
plt.show()
```

Los colores se ven diferentes porque OpenCV carga las imágenes en formato BGR pero Matplotlib asume que las imágenes están en RGB Al no hacer la conversión, los canales se muestran invertidos.
