## Funcion

```python
def data_img(img):
  print('Tamaño = ',img.shape)
  print('Max  = ',np.max(img))
  print('Min  = ',np.min(img))

data_img(img)
```

## Recortar

```python
# Primero alto, despues ancho
img_cropped = img[80:250, 200:420]
plt.imshow(img_cropped)
```

```python
data_img(img_cropped)
```

```python
img_canny = cv2.Canny(img_cropped, 150, 250)
plt.imshow(img_canny)
```

```python
# Primero ancho y despues alto
img_resize = cv2.resize(img_cropped, (60, 50))
plt.imshow(img_resize)
```

```python
# Apilado vertical
ver = np.vstack((img_resize, img_resize))
plt.imshow(ver);
```

```python
# Apilamos horizontamente dos filas, generando una "matriz" de 2x2
hor2 = np.hstack((ver, ver))
plt.imshow(hor2);
```
